## Task 1.4.1
First assign to the variable S a list or set consisting of the complex numbers listed as S = {2 + 2i, 3 + 2i, 1.75 + 1i, 2 + 1i, 2.25 +
1i, 2.5 + 1i, 2.75 + 1i, 3 + 1i, 3.25 + 1i}

```python
import matplotlib.pyplot as plt

def scatter_plot(data, marker_size, xlim, ylim):
    """
    Creates a scatter plot of complex numbers.

    Args:
        data: A set of complex numbers.
        marker_size: The size of the markers in the scatter plot.
    """

    # Extract real and imaginary parts
    x = [num.real for num in data]
    y = [num.imag for num in data]

    # Create the scatter plot
    plt.scatter(x, y, s=marker_size**2)  # Square marker_size for area scaling
    plt.xlim(left=0, right=xlim) # Set only the left x limit
    plt.ylim(top=ylim) # Set only the top y limit

    # Add labels and title
    plt.xlabel("Real")
    plt.ylabel("Imaginary")
    plt.title("Scatter Plot of Complex Numbers")

    # Show the plot
    plt.show()
```

```python
S = {2 + 2j, 3 + 2j, 1.75 + 1j, 2 + 1j, 2.25 + 1j, 2.5 + 1j, 2.75 + 1j, 3 + 1j, 3.25 + 1j}
scatter_plot(S, 4) # Call the function with your data and marker size
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the starting points of the vectors
origins = np.array([[0, 0], [0, 0], [0, 0]])

# Define the vectors as displacements from the origin
vectors = np.array([[2, 3], [3, -2], [5, 1]])

# Separate the x and y components of the vectors
x = vectors[:, 0]
y = vectors[:, 1]

# Use quiver to plot the vectors
plt.quiver(origins[:, 0], origins[:, 1], x, y, angles='xy', scale_units='xy', scale=1)

# Set plot limits and labels
plt.xlim(-1, 6)
plt.ylim(-3, 4)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Plot of Three Vectors')
plt.grid(True)

# Show the plot
plt.show()
```

# Problem 1.4.6
Draw a diagram representing the complex number $z_0$ = -3+3i using two arrows with their tails located at different points.

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the starting points of the vectors
origins = np.array([[0, 0], [0, 3], [0, 0]])

# Define the vectors as displacements from the origin
vectors = np.array([[0, 3], [-3, 0], [-3, 3]])

# Separate the x and y components of the vectors
x = vectors[:, 0]
y = vectors[:, 1]

# Use quiver to plot the vectors
plt.quiver(origins[:, 0], origins[:, 1], x, y, angles='xy', scale_units='xy', scale=1)

# Set plot limits and labels
plt.xlim(-4, 6)
plt.ylim(-3, 4)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Plot of Three Vectors')
plt.grid(True)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

def scatter_plot(data, marker_size, xlim, ylim):
    """
    Creates a scatter plot of complex numbers.

    Args:
        data: A set of complex numbers.
        marker_size: The size of the markers in the scatter plot.
    """

    # Extract real and imaginary parts
    x = [num.real for num in data]
    y = [num.imag for num in data]

    # Create the scatter plot
    plt.scatter(x, y, s=marker_size**2)  # Square marker_size for area scaling
    # Draw lines at x=0 and y=0
    plt.axhline(y=0, color='black', linestyle='--')
    plt.axvline(x=0, color='black', linestyle='--')

    plt.xlim(left=-4, right=xlim) # Set only the left x limit
    plt.ylim(top=ylim, bottom=-4) # Set only the top y limit

    # Add labels and title
    plt.xlabel("Real")
    plt.ylabel("Imaginary")
    plt.title("Scatter Plot of Complex Numbers")

    # Show the plot
    plt.show()
```

```python
def _flat2boxed(row):
    # Notw we skip every 4th element, thus eliminating the alpha channel.
    return [tuple(row[i:i+3]) for i in range(0, len(row), 4)]
    
```

```python
def file2image(path):
    """Reads an image into list of lists of pixel values (tuples with three values). This is a color image."""
    (w, h, p, m) = png.Reader(filename=path).asRGBA()
    return [_flat2boxed(r) for r in p]
```
