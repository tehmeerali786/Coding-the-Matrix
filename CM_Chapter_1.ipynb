```python
import matplotlib.pyplot as plt

def scatter_plot(data, marker_size, xlim, ylim):
    """
    Creates a scatter plot of complex numbers.

    Args:
        data: A set of complex numbers.
        marker_size: The size of the markers in the scatter plot.
    """

    # Extract real and imaginary parts
    x = [num.real for num in data]
    y = [num.imag for num in data]

    # Create the scatter plot
    plt.scatter(x, y, s=marker_size**2)  # Square marker_size for area scaling
    plt.xlim(left=0, right=xlim) # Set only the left x limit
    plt.ylim(top=ylim) # Set only the top y limit

    # Add labels and title
    plt.xlabel("Real")
    plt.ylabel("Imaginary")
    plt.title("Scatter Plot of Complex Numbers")

    # Show the plot
    plt.show()
```

```python
S = {2 + 2j, 3 + 2j, 1.75 + 1j, 2 + 1j, 2.25 + 1j, 2.5 + 1j, 2.75 + 1j, 3 + 1j, 3.25 + 1j}
scatter_plot(S, 4) # Call the function with your data and marker size
```

```python
S_scaled = {1+2j+z for z in S}
```

```python
scatter_plot(S_scaled, 4, 6, 6)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Single vector using plt.arrow()
plt.figure()
plt.arrow(0, 0, 3, 4, head_width=0.2, head_length=0.3, fc='blue', ec='black')
plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Single Vector using plt.arrow()')
plt.grid(True)
plt.show()

# Multiple vectors using plt.quiver()
origin = np.array([[0, 0], [0, 0]]) # origin point for each vector
vectors = np.array([[2, 2], [3, -1]]) # vector coordinates

plt.figure()
plt.quiver(*origin, vectors[:,0], vectors[:,1], color=['r','g'], scale=10)
plt.xlim(-1, 4)
plt.ylim(-2, 3)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Multiple Vectors using plt.quiver()')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the starting points of the vectors
origins = np.array([[0, 0], [0, 0], [0, 0]])

# Define the vectors as displacements from the origin
vectors = np.array([[2, 3], [3, -2], [5, 1]])

# Separate the x and y components of the vectors
x = vectors[:, 0]
y = vectors[:, 1]

# Use quiver to plot the vectors
plt.quiver(origins[:, 0], origins[:, 1], x, y, angles='xy', scale_units='xy', scale=1)

# Set plot limits and labels
plt.xlim(-1, 6)
plt.ylim(-3, 4)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Plot of Three Vectors')
plt.grid(True)

# Show the plot
plt.show()
```
